```r
retaildata <- readxl::read_excel("retail.xlsx", skip=1)
myts <- ts(retaildata[,"A3349873A"],
  frequency=12, start=c(1982,4))
myts.train <- window(myts, end=c(2010,12))
myts.test <- window(myts, start=2011)

autoplot(myts) +
  autolayer(myts.train, series="Training") +
  autolayer(myts.test, series="Test")
```

```r
fc <- snaive(myts.train)
accuracy(fc,myts.test)
```

```r
checkresiduals(fc)
```

```r
autoplot(myts) +
    autolayer(fc, series = "forecast") +
    autolayer(myts.test, series = "test set")
```

```r
help(visnights)
# Quarterly visitor nights
autoplot(visnights)
```

```r
train1 <- window(visnights[, "QLDMetro"], end = c(2015, 4))
train2 <- window(visnights[, "QLDMetro"], end = c(2014, 4))
train3 <- window(visnights[, "QLDMetro"], end = c(2013, 4))
fc1 <- snaive(train1, h = 4)
fc2 <- snaive(train2, h = 4)
fc3 <- snaive(train3, h = 4)
```

```r
accuracy(fc1, window(visnights[, "QLDMetro"], start = c(2016, 1)))
accuracy(fc2, window(visnights[, "QLDMetro"], start = c(2015, 1), end = c(2015, 4)))
accuracy(fc3, window(visnights[, "QLDMetro"], start = c(2014, 1), end = c(2014, 4)))
```

```r
autoplot(visnights[, "QLDMetro"]) +
    autolayer(fc1, series = "forecast 1") +
    autolayer(fc2, series = "forecast 2") +
    autolayer(fc3, series = "forecast 3") +
    autolayer(window(visnights[, "QLDMetro"], start = c(2014, 1)), series = "test")
```

MAPE(fc2) < MAPE(fc1) < MAPE(fc3)

```r
autoplot(dowjones)
```

```r
fc <- rwf(dowjones, drift = T)
autoplot(fc)
```

```r
drift <- rwf(dowjones, drift = T)
naive <- naive(dowjones)
mean <- meanf(dowjones)
autoplot(dowjones) +
    autolayer(drift, series = "drift fc", PI = F) +
    autolayer(naive, series = "naive fc", PI = F) +
    autolayer(mean, series = "mean fc", PI = F)
```

```r
autoplot(ibmclose)
```

```r
train <- window(ibmclose, end = 300)
test <- window(ibmclose, start = 300)
```

```r
fc1 <- naive(train, h = 69)
fc2 <- rwf(train, drift = T, h = 69)
fc3 <- meanf(train, h = 69)
```

```r
autoplot(ibmclose) + 
    autolayer(fc1, series = "naive fc", PI = F) +
    autolayer(fc2, series = "drift fc", PI = F) +
    autolayer(fc3, series = "mean fc", PI = F) +
    autolayer(test, series = "real data")
```

```r
accuracy(fc1, test)
accuracy(fc2, test)
accuracy(fc3, test)
```

Drift method makes the best forecast

```r
checkresiduals(fc2)
```

The residuals from drift roughly resemble white noise

```r
help(hsales)
# Monthly sales of new one-family houses sold in the USA since 1973.
autoplot(hsales)
```

```r
train <- window(hsales, end = c(1993, 12))
test <- window(hsales, start = c(1994, 1))
```

```r
naive_fc <- naive(train, h = 23)
drift_fc <- rwf(train, drift = T, h = 23)
mean_fc <- meanf(train, h = 23)
snaive_fc <- snaive(train, h = 23)
```

```r
autoplot(hsales) +
    autolayer(naive_fc, series = "naive fc", PI = F) +
    autolayer(drift_fc, series = "drift fc", PI = F) +
    autolayer(mean_fc, series = "mean fc", PI = F) +
    autolayer(snaive_fc, series = "snaive fc", PI = F) +
    autolayer(test, series = "test data")
```

```r
accuracy(naive_fc, test)
accuracy(drift_fc, test)
accuracy(mean_fc, test)
accuracy(snaive_fc, test)
```

```r
checkresiduals(snaive_fc)
```

Seasonal naive method makes the best forecast, but its residuals are not white noise.
